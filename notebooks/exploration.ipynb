# Phishing Detection - Data Exploration

Exploratory analysis of the phishing detection dataset.

```python
import pandas as pd
import numpy as np
from pathlib import Path

DATA_DIR = Path("../data/processed")
legit = pd.read_csv(DATA_DIR / "legitimate.csv")
phish = pd.read_csv(DATA_DIR / "phishing.csv")
df = pd.concat([legit, phish], ignore_index=True)
print(f"Shape: {df.shape}")
print(f"Class distribution:\n{df['Label'].value_counts()}")
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

feature_cols = [c for c in df.columns if c not in ['Domain', 'Label']]
fig, axes = plt.subplots(4, 4, figsize=(14, 12))
axes = axes.flatten()
for i, col in enumerate(feature_cols[:16]):
    df.groupby('Label')[col].value_counts().unstack(fill_value=0).plot(kind='bar', ax=axes[i], legend=False)
    axes[i].set_title(col)
plt.tight_layout()
plt.show()
```
